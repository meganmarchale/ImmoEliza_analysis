Histogram bedrooms and price ranges for all Belgium

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(input_file)

price_bins = [0, 100000, 200000, 300000, 400000, 500000, 600000, 700000, 800000, float('inf')]
price_labels = [
    "0-100k", "100k-200k", "200k-300k", "300k-400k",
    "400k-500k", "500k-600k", "600k-700k", "700k-800k", "800k+"
]
df['price_range'] = pd.cut(df['price'], bins=price_bins, labels=price_labels, right=True)

df['bedroomCount'] = pd.to_numeric(df['bedroomCount'], errors='coerce')
df['bedroomCount'] = df['bedroomCount'].apply(lambda x: int(x) if x < 5 else 5)

bedroom_labels = ['0', '1', '2', '3', '4', '5+']
df['bedroomGroup'] = df['bedroomCount'].astype(str)
df['bedroomGroup'] = df['bedroomGroup'].replace({'5': '5+'})

pivot = pd.crosstab(df['price_range'], df['bedroomGroup'])

ax = pivot.plot(kind='bar', stacked=True, figsize=(12, 6), colormap='tab10')


totals = pivot.sum(axis=1)

for i, total in enumerate(totals):
    ax.text(i, total + 1, str(total), ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.title("Number of properties by range of price and number of bedrooms")
plt.xlabel("Price ranges")
plt.ylabel("Number of properties")
plt.legend(title="Bedrooms")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Price ranges houses / apartments

```python
import pandas as pd
import plotly.express as px

# Charger les données
df = pd.read_csv(input_file)

# Nettoyage
df = df.copy()
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df = df[df['region'] == 'Brussels']
df = df[df['type'].isin(['HOUSE', 'APARTMENT'])]
df = df.dropna(subset=['price'])

# Vérifie la distribution des types
print("Types disponibles :", df['type'].value_counts())
print("Nombre de lignes après filtrage :", len(df))

# Créer les tranches de prix
price_bins = [0, 100000, 200000, 300000, 400000, 500000, 600000, 700000, 800000, float('inf')]
price_labels = [
    "0-100k", "100k-200k", "200k-300k", "300k-400k",
    "400k-500k", "500k-600k", "600k-700k", "700k-800k", "800k+"
]
df['price_range'] = pd.cut(df['price'], bins=price_bins, labels=price_labels, right=True)

# Vérifie s’il y a des NaN dans price_range
print("Valeurs manquantes dans price_range :", df['price_range'].isna().sum())

# Groupement
grouped = df.groupby(['price_range', 'type']).size().reset_index(name='count')
print(grouped)

# Graphique
fig = px.bar(
    grouped,
    x='price_range',
    y='count',
    color='type',
    title="Houses and apartments per price range",
    labels={'price_range': 'Price range', 'count': 'Number of properties', 'type': 'Type'},
    barmode='stack',
    text='count'
)

fig.update_layout(xaxis_tickangle=-45)
fig.show()
```
